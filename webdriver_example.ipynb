```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.remote.webelement import WebElement

import bs4
from IPython import display
from PIL import Image
```

### Define parameters

```python
window_width = 1380
window_height = 720
timeout = 10
screenshot_filename = 'screenshot.png'
screenshot_path = os.path.join(os.getcwd(), screenshot_filename)
test_url = 'https://www.google.com'
```

### Define WebDriver client

```python
def get_webdriver(timeout, window_width,window_height):
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--start-maximized')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  # added for timeout
  options.add_argument('--disable-gpu')

  # initialize
  driver = webdriver.Chrome(options=options)

  # Set client parameters
  driver.implicitly_wait(5)
  driver.set_page_load_timeout(timeout)

  driver.set_window_size(window_width, window_height)

  driver.execute_script(f'window.innerWidth = {window_width}')
  driver.execute_script(f'window.innerHeight = {window_height}')

  return driver
```

```python
driver = get_webdriver(timeout, window_width,window_height)
```

```python
print('screenshot path is', screenshot_path)
```

### Use WebDriver to navigate to the URL and save a screenshot locally

```python

driver.get(test_url)
# get a screenshot at the destination of the URL and save the file locally
saved_screenshot = driver.save_screenshot(screenshot_path)
```
